```python
# Colab: Multi-ticker ARMA/ARIMA, Unit Root, Structural Break, Cointegration, Random Walk

!pip install arch

import pandas as pd
import numpy as np
from google.colab import files
import matplotlib.pyplot as plt

from statsmodels.tsa.stattools import adfuller, zivot_andrews, coint
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from arch.unitroot import PhillipsPerron

# --- Tickers and Excel file mapping ---
TICKERS = ["AAL", "UAL", "JBLU", "XOM", "WMT", "AMZN"]
price_col = "Close"
date_col = "Date"

print(f"Please upload Excel files for: {', '.join(TICKERS)} (one per ticker)")
uploaded = files.upload()
# Map uploaded filenames to tickers (by substring match)
ticker_files = {}
for ticker in TICKERS:
    match = [fname for fname in uploaded.keys() if ticker in fname.upper()]
    if match:
        ticker_files[ticker] = match[0]
    else:
        print(f"WARNING: No file found for {ticker}")

# --- Load DataFrames ---
dfs = {}
for ticker, fname in ticker_files.items():
    try:
        df = pd.read_excel(fname)
        dfs[ticker] = df
    except Exception as e:
        print(f"Error loading {ticker} ({fname}): {e}")

# --- Extract Closing Price Series ---
close_series = {}
date_series = {}
for ticker, df in dfs.items():
    try:
        close_series[ticker] = df[price_col].dropna().reset_index(drop=True)
        date_series[ticker] = df[date_col] if date_col in df.columns else None
        print(f"{ticker}: {len(close_series[ticker])} closing prices found.")
    except KeyError:
        print(f"Error: '{price_col}' column not found in {ticker}")

# --- Unit Root Tests ---
def adf_test(series, name):
    print(f"\nADF Test for {name}:")
    result = adfuller(series, regression='c', autolag='AIC')
    print(f"ADF Statistic: {result[0]:.4f}, p-value: {result[1]:.4f}")
    print("Interpretation:", "Stationary" if result[1] < 0.05 else "Non-stationary")
    return result

def pp_test(series, name):
    print(f"\nPhillips-Perron Test for {name}:")
    result = PhillipsPerron(series, lags=None)
    print(f"PP Statistic: {result.stat:.4f}, p-value: {result.pvalue:.4f}")
    print("Interpretation:", "Stationary" if result.pvalue < 0.05 else "Non-stationary")
    return result

def zivot_andrews_test(series, name):
    print(f"\nZivot-Andrews Test for {name}:")
    result = zivot_andrews(series, regression='c', autolag='AIC')
    print(f"ZA Statistic: {result[0]:.4f}, p-value: {result[1]:.4f}, Breakpoint: {result[3]}")
    print("Interpretation:", "Structural break detected" if result[1] < 0.05 else "No clear break")
    return result

# --- ARMA/ARIMA/Random Walk Modeling ---
def fit_arima(series, name):
    print(f"\nFitting ARIMA(1,1,1) for {name}:")
    try:
        model = ARIMA(series, order=(1,1,1))
        res = model.fit()
        print(res.summary())
        forecast = res.forecast(steps=1)[0]
        print(f"One-step ahead forecast for {name} using ARIMA: {forecast:.4f}")
        return res, forecast
    except Exception as e:
        print(f"ARIMA fit error for {name}: {e}")
        return None, None

def fit_arma(series, name):
    print(f"\nFitting ARMA(2,0) for {name}:")
    try:
        model = AutoReg(series, lags=2, old_names=False)
        res = model.fit()
        print(res.summary())
        forecast = res.forecast(steps=1)[0]
        print(f"One-step ahead forecast for {name} using ARMA: {forecast:.4f}")
        return res, forecast
    except Exception as e:
        print(f"ARMA fit error for {name}: {e}")
        return None, None

def fit_random_walk(series, name):
    print(f"\nFitting Random Walk (ARIMA(0,1,0)) for {name}:")
    try:
        model = ARIMA(series, order=(0,1,0))
        res = model.fit()
        print(res.summary())
        forecast = res.forecast(steps=1)[0]
        print(f"One-step ahead forecast for {name} using Random Walk: {forecast:.4f}")
        return res, forecast
    except Exception as e:
        print(f"Random Walk fit error for {name}: {e}")
        return None, None

# --- Plot series and breakpoints ---
def plot_series_with_breakpoint(series, name, breakpoint_idx, dates=None):
    plt.figure(figsize=(12, 6))
    plt.plot(series, label=f'{name} Closing')
    plt.axvline(x=breakpoint_idx, color='red', linestyle='--', label=f'ZA Breakpoint {breakpoint_idx}')
    plt.title(f'{name} Closing Prices and Structural Break')
    plt.xlabel('Index' if dates is None else 'Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()
    if dates is not None and 0 <= breakpoint_idx < len(dates):
        print(f"{name} break date: {dates.iloc[breakpoint_idx]}")
    else:
        print(f"{name} break index: {breakpoint_idx}")

# --- Run unit root, break, models ---
results = {}
for ticker, series in close_series.items():
    print(f"\n--- {ticker} Analysis ---")
    adf = adf_test(series, ticker)
    pp = pp_test(series, ticker)
    za = zivot_andrews_test(series, ticker)
    plot_series_with_breakpoint(series, ticker, za[3], date_series[ticker])
    arma_res, arma_forecast = fit_arma(series, ticker)
    arima_res, arima_forecast = fit_arima(series, ticker)
    rw_res, rw_forecast = fit_random_walk(series, ticker)
    results[ticker] = {
        "adf": adf,
        "pp": pp,
        "za": za,
        "arma": arma_res,
        "arma_forecast": arma_forecast,
        "arima": arima_res,
        "arima_forecast": arima_forecast,
        "random_walk": rw_res,
        "rw_forecast": rw_forecast,
    }

# --- Cointegration Tests between all pairs ---
def cointegration_test(series1, series2, label1, label2):
    print(f"\nCointegration Test: {label1} vs {label2}")
    coint_t, pvalue, crit = coint(series1, series2)
    print(f"Statistic: {coint_t:.4f}, p-value: {pvalue:.4f}")
    print("Interpretation:", "Cointegrated" if pvalue < 0.05 else "Not cointegrated")
    return (coint_t, pvalue, crit)

from itertools import combinations
for t1, t2 in combinations(close_series.keys(), 2):
    series1 = close_series[t1]
    series2 = close_series[t2]
    min_len = min(len(series1), len(series2))
    cointegration_test(series1[:min_len], series2[:min_len], t1, t2)
```
